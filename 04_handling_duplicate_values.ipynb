# Handling Duplicate Values

```python
import pandas as pd
```

```python
data = pd.read_excel("duplicate_values.xlsx")
data
```

#### To find duplicate values.

```python
data["Customer_Name"].duplicated().sum()
```

#### To drop duplicate values.

```python
data.drop_duplicates()
```

To remove duplicates by column.

```python
data2 = data.drop_duplicates("Order_Number")
data2
```
