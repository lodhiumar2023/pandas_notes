# Merge
### Merging two data set on the basis of primary key

```python
import pandas as pd

data1 = {
    "id" : [1, 2, 3, 4, 5, 6],
    'name' : ['Kashif', 'Nasir', 'Ahmed', 'Saleem', 'Javed', 'Rashid'],
    'age' : [30, 25, 21, 32, 29, 28]
}

data2 = {
    "id" : [1, 3, 2, 4, 5, 6],
    'salary' : [30000, 50000, 20000, 35000, 25000, 21000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)



merge_data = pd.merge(df1, df2, on = 'id')
merge_data
```

## Merging data by left or right columns

#### Right Merge

```python
data1 = {
    "id" : [1, 2, 3, 4, 5, 6],
    'name' : ['Kashif', 'Nasir', 'Ahmed', 'Saleem', 'Javed', 'Rashid'],
    'age' : [30, 25, 21, 32, 29, 28]
}

data2 = {
    "id" : [1, 3, 2, 4, 7, 8],
    'salary' : [30000, 50000, 20000, 35000, 25000, 21000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)


right_merge = pd.merge(left = df1, right = df2, on = 'id', how = 'right')


right_merge
```

#### Left Merge

```python
data1 = {
    "id" : [1, 2, 3, 4, 5, 6],
    'name' : ['Kashif', 'Nasir', 'Ahmed', 'Saleem', 'Javed', 'Rashid'],
    'age' : [30, 25, 21, 32, 29, 28]
}

data2 = {
    "id" : [1, 3, 2, 4, 7, 8],
    'salary' : [30000, 50000, 20000, 35000, 25000, 21000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

left_merge = pd.merge(left = df1, right = df2, on = 'id', how = 'left')

left_merge
```

#### Concatenation of two data sets

```python
data1 = {
    "id" : [1, 2, 3, 4, 5, 6],
    'name' : ['A', 'B', 'C', 'D', 'F', 'G'],
    'age' : [30, 25, 21, 32, 29, 28]
}

data2 = {
    "id" : [7, 8, 9, 10, 11, 12],
    'name' : ['H', 'I', 'J', 'K', 'L', 'M'],
    'age' : [30, 25, 21, 32, 29, 28]
}

df1 = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

data_con = pd.concat([df1, df2])
data_con
```
