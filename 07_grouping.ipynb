# Grouping Data

```python
import pandas as pd
```

```python
data = pd.read_excel('DataSet.xlsx')
data = data.dropna()
data.info()
```

### Groupby using agg count, in groupby and agg one or more parameters can be given

```python
gp = data.groupby('Status').agg({'Status': 'count'})
gp
```

```python
gp = data.groupby(['Status', 'Category']).agg({'Status': 'count'})
gp
```

### Groupby using agg mean

```python
gp1 = data.groupby(['Status']).agg({'Sales':'mean'})
gp1
```

```python
gp1 = data.groupby(['Status']).agg({'Quantity':'count', 'Total_Cost': 'sum', 'Total_Sales':'sum'})
gp1
```
